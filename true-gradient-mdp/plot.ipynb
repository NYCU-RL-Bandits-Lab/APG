```python
import os
import yaml
import argparse
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import Image, display

from helper import plot, configure
```

### Fig 1. Sigmoid Function

```python
fig = plt.figure(figsize=(5, 4))

# configuration
fontsize = 18
linewidth = 3.0
axis = plt.subplot(1, 1, 1)

# plot
ra1 = np.arange(-5, 0.2, 0.1)
ra2 = np.arange(0, 5, 0.1)
axis.plot(
    ra1, [1/(1+np.exp(-x)) for x in ra1], 
    color="coral", 
    linewidth=linewidth,
    linestyle='--'
)
axis.plot(
    ra2, 
    [1/(1+np.exp(-x)) for x in ra2], 
    color="seagreen",
    linewidth=linewidth,
    linestyle='-',
)

plt.text(1.8, 0.73, 'Concave', fontsize=fontsize, fontdict=dict(weight='bold'), fontname='monospace', color="seagreen")
plt.text(-4.5, 0.17, 'Convex', fontsize=fontsize, fontdict=dict(weight='bold'), fontname='monospace', color="coral")

axis.set_title("Sigmoid Function", fontsize=fontsize, fontdict=dict(weight='bold'), fontname='monospace', pad=12)
axis.set_xlabel("Θ($a^*$) - Θ($a_2$)", fontsize=fontsize, fontname='monospace', labelpad=7)
axis.set_ylabel("$V(s)$", fontsize=fontsize, fontname='monospace', labelpad=7)

# configuration 
configure(
    axis,
    legendsize=None,
    grid=False,
)
plt.xticks([-4, -2, 0, 2, 4])

# save plot
plt.tight_layout()  
plt.savefig('./logs/sigmoid.png')
display(Image(filename='./logs/sigmoid.png'))
```
